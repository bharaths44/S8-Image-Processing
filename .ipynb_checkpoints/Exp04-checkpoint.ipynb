```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image in grayscale
image = cv2.imread('C:/Users/Student/Pictures/image.jpg', cv2.IMREAD_GRAYSCALE)

# Normalize the image to range [0, 1]
image_normalized = image/255.0
```

```python
# Transformation 1: Contrast and brightness adjustment (s = a*r + b)
a = 1.5  # Scaling factor for contrast
b = 0.2  # Brightness adjustment
transformed_contrast_brightness = a * image_normalized + b
transformed_contrast_brightness = np.clip(transformed_contrast_brightness, 0, 1)  # Ensure in range [0,1]
transformed_contrast_brightness = (transformed_contrast_brightness * 255).astype(np.uint8)

# Transformation 2: Logarithmic transformation (s = c * log(1 + r))
c = 1.0  # Scaling constant
transformed_logarithmic = c * np.log1p(image_normalized)
transformed_logarithmic = np.clip(transformed_logarithmic, 0, 1)  # Ensure in range [0,1]
transformed_logarithmic = (transformed_logarithmic * 255).astype(np.uint8)

# Transformation 3: Gamma correction (s = c * r^gamma)
gamma = 0.5  # Gamma < 1 brightens the image
transformed_gamma = np.power(image_normalized, gamma)
transformed_gamma = np.clip(transformed_gamma, 0, 1)  # Ensure in range [0,1]
transformed_gamma = (transformed_gamma * 255).astype(np.uint8)
```

```python
# Display the original and transformed images
plt.figure(figsize=(12, 6))

# Original image
plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Contrast and brightness adjusted
plt.subplot(1, 4, 2)
plt.imshow(transformed_contrast_brightness, cmap='gray')
plt.title('Contrast & Brightness')
plt.axis('off')

# Logarithmic transformation
plt.subplot(1, 4, 3)
plt.imshow(transformed_logarithmic, cmap='gray')
plt.title('Logarithmic Transformation')
plt.axis('off')

# Gamma corrected image
plt.subplot(1, 4, 4)
plt.imshow(transformed_gamma, cmap='gray')
plt.title('Gamma Correction')
plt.axis('off')

plt.show()
```
