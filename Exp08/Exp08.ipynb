```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read and Display the Input Image
def load_image(image_path):
    """Load the image in grayscale."""
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print("Error: Image not found!")
        return None
    return img

# Step 2: Apply Smoothing (Blurring) Filters
def apply_smoothing_filters(img):
    """Apply different smoothing filters."""
    # Mean (Averaging) Filter
    mean_filter = cv2.blur(img, (5, 5))  # 5x5 kernel
    
    # Gaussian Filter
    gaussian_filter = cv2.GaussianBlur(img, (5, 5), 0)
    
    # Median Filter
    median_filter = cv2.medianBlur(img, 5)  # 5x5 kernel
    
    return mean_filter, gaussian_filter, median_filter

# Step 3: Apply Sharpening Filters
def apply_sharpening_filters(img):
    """Apply different sharpening filters."""
    # Laplacian Filter (Edge detection)
    laplacian_filter = cv2.Laplacian(img, cv2.CV_64F)
    laplacian_filter = cv2.convertScaleAbs(laplacian_filter)  # Convert back to uint8
    
    # Unsharp Masking
    blurred_img = cv2.GaussianBlur(img, (5, 5), 0)
    unsharp_masking = cv2.addWeighted(img, 1.5, blurred_img, -0.5, 0)
    
    return laplacian_filter, unsharp_masking

# Step 4: Display and Compare Results
def compare_results(original, mean, gaussian, median, laplacian, unsharp):
    """Display original and processed images."""
    plt.figure(figsize=(12, 6))

    # Original Image
    plt.subplot(2, 3, 1)
    plt.imshow(original, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')

    # Mean Filter Result
    plt.subplot(2, 3, 2)
    plt.imshow(mean, cmap='gray')
    plt.title("Mean Filter (Averaging)")
    plt.axis('off')

    # Gaussian Filter Result
    plt.subplot(2, 3, 3)
    plt.imshow(gaussian, cmap='gray')
    plt.title("Gaussian Filter")
    plt.axis('off')

    # Median Filter Result
    plt.subplot(2, 3, 4)
    plt.imshow(median, cmap='gray')
    plt.title("Median Filter")
    plt.axis('off')

    # Laplacian Filter Result
    plt.subplot(2, 3, 5)
    plt.imshow(laplacian, cmap='gray')
    plt.title("Laplacian Filter (Sharpening)")
    plt.axis('off')

    # Unsharp Masking Result
    plt.subplot(2, 3, 6)
    plt.imshow(unsharp, cmap='gray')
    plt.title("Unsharp Masking")
    plt.axis('off')

    plt.show()

# Main Function
def main(image_path):
    # Step 1: Load Image
    img = load_image(image_path)
    if img is None:
        return

    # Step 2: Apply Smoothing Filters
    mean_filter, gaussian_filter, median_filter = apply_smoothing_filters(img)

    # Step 3: Apply Sharpening Filters
    laplacian_filter, unsharp_masking = apply_sharpening_filters(img)

    # Step 4: Display and Compare the Results
    compare_results(img, mean_filter, gaussian_filter, median_filter, laplacian_filter, unsharp_masking)

# Run the Program
if __name__ == "__main__":
    image_path = 'C:/Users/Student/Documents/Img processing/image.jpg'  # Replace with your image file path
    main(image_path)
```
