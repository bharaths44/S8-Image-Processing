```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the negative of a grayscale image
def negative_grayscale(image):
    return 255 - image

# Function to compute the binary negative of an image
def negative_binary(image, threshold=127):
    # Convert image to binary using the given threshold
    _, binary_image = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)
    # Compute binary negative
    return 255 - binary_image

# Main function to simulate and display the image, its negative (grayscale and binary)
def main():
    # Load the original image in grayscale
    image_path = "C:/Users/Student/Pictures/image.jpg"

    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if image is None:
        print(f"Error: Could not load image at {image_path}")
        return
    
    # Grayscale Negative
    grayscale_negative = negative_grayscale(image)
    
    # Binary Negative
    binary_negative = negative_binary(image)
    
    # Displaying the results
    plt.figure(figsize=(12, 4))
    
    # Original Image
    plt.subplot(1, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    
    # Grayscale Negative Image
    plt.subplot(1, 3, 2)
    plt.imshow(grayscale_negative, cmap='gray')
    plt.title('Grayscale Negative')
    plt.axis('off')
    
    # Binary Negative Image
    plt.subplot(1, 3, 3)
    plt.imshow(binary_negative, cmap='gray')
    plt.title('Binary Negative')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Run the main function
if __name__ == "__main__":
    main()
```
