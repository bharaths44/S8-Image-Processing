```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 1. Image Sampling
def image_sampling(I, F):
    """
    Reduces the resolution of the image by skipping rows and columns.

    Parameters:
        I (numpy array): Original grayscale image.
        F (int): Sampling factor.

    Returns:
        Is (numpy array): Sampled image.
    """
    # Initialize the sampled image
    Is = I[::F, ::F]
    
    return Is

# 2. Image Quantization
def image_quantization(I, Q):
    """
    Reduces the number of intensity levels in the image.

    Parameters:
        I (numpy array): Original grayscale image.
        Q (int): Number of quantization levels.

    Returns:
        Iq (numpy array): Quantized image.
    """
    # Normalize the pixel values to [0, 1]
    I_prime = I / 255.0
    
    # Quantize the normalized values to range [0, Q-1]
    Iq_prime = np.round(I_prime * (Q - 1))
    
    # Scale back to range [0, 255]
    Iq = (Iq_prime * (255 / (Q - 1))).astype(np.uint8)
    
    return Iq

# Main function to execute both algorithms
def main():
    # Load grayscale image
    image_path = "C:/Users/Student/Pictures/image.jpg"
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if I is None:
        print(f"Error: Could not load image at {image_path}")
        return
    
    # 1. Image Sampling
    F = 2  # Sampling factor (you can change this)
    Is = image_sampling(I, F)

    # 2. Image Quantization
    Q = 4  # Number of quantization levels (you can change this)
    Iq = image_quantization(I, Q)

    # Display original, sampled, and quantized images using Matplotlib
    plt.figure(figsize=(12, 4))

    # Original Image
    plt.subplot(1, 3, 1)
    plt.imshow(I, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Sampled Image
    plt.subplot(1, 3, 2)
    plt.imshow(Is, cmap='gray')
    plt.title('Sampled Image')
    plt.axis('off')

    # Quantized Image
    plt.subplot(1, 3, 3)
    plt.imshow(Iq, cmap='gray')
    plt.title('Quantized Image')
    plt.axis('off')

    # Show the images
    plt.tight_layout()
    plt.show()

# Run the main function
if __name__ == "__main__":
    main()
```
