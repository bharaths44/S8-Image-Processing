```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the image
image = cv2.imread('C:/Users/Student/Pictures/image.jpg', cv2.IMREAD_GRAYSCALE)

# Step 2: Apply Discrete Fourier Transform (DFT)
dft = np.fft.fft2(image)

# Step 3: Shift zero frequency to the center
dft_shifted = np.fft.fftshift(dft)

# Step 4: Compute the magnitude spectrum
magnitude_spectrum = np.abs(dft_shifted)
magnitude_spectrum = np.log(1 + magnitude_spectrum)  # Log scale for better visualization
```

```python
# Step 5: Display the results
plt.figure(figsize=(12, 6))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Magnitude Spectrum
plt.subplot(1, 2, 2)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum')
plt.axis('off')

plt.show()
```

```python

# Optional: Inverse DFT
# Step 6: Revert the frequency shift
dft_shifted_back = np.fft.ifftshift(dft_shifted)

# Apply inverse DFT to get the image back
image_reconstructed = np.fft.ifft2(dft_shifted_back)

# Take the real part (since the result can have a very small imaginary part)
image_reconstructed = np.real(image_reconstructed)

# Display the reconstructed image
plt.figure(figsize=(6, 6))
plt.imshow(image_reconstructed, cmap='gray')
plt.title('Reconstructed Image from Inverse DFT')
plt.axis('off')
plt.show()
```
